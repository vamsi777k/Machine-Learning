```python
import numpy as np
import math 
from math import sqrt
 

def clean_data(line):
    return line.replace('(', '').replace(')', '').replace(' ', '').strip().split(',')

def fetch_data(filename):
    with open(filename, 'r') as f:
        input_data = f.readlines()
        clean_input = list(map(clean_data, input_data))
        f.close()
    return clean_input


def readFile(dataset_path):
    input_data = fetch_data(dataset_path)
    input_np = np.array(input_data)
    return input_np

training_data = './dataset/1a-training.txt'
test_data = './dataset/1a-test.txt'
large_120_data = './dataset/1c-data.txt'

train_np = readFile(training_data)
print(train_np)
test_np = readFile(test_data)
print(test_np)
large_np = readFile(large_120_data)
```

```python
#deriving the parameters for the gaussian naive bayes calssifier.

def check(a):                #defining check
    vamsi = {'W':0,'M':0 }   #assign men and women as zero
    for GVK in a:            #for gvk in a
        if GVK=='W':         #if gvk equal to women
            vamsi['W'] += 1  #then vamsi of women is imncreasing if its equal to one
        if GVK=='M':         #if gvk equal to men
            vamsi['M'] += 1  #then vamsi of men is increasing if its equal to one
    return vamsi             #return vamsi
#by using cartesian distance for immediate steps of generating training data sets.        
given = [1,3,7]        #generated training data for values of given of 1, 3, and 7.
for l in given:        #for l in given then
    print("l : value ", l)     #print l value
    for wintest in large_np:    #for w in test in given large_np data set
        length = []             #declaring length                      
        for railu in  train_np:    #for railu in given train_np data set
            q = np.array( [float(a) for a in railu[:3]])
            w = np.array( [float(a) for a in wintest[:3]])
            e = sqrt(np.sum( (q-w)**2 ))   #using squareroot function and calculating it
            length.append(e)      #appending e and calculating length of it
        
        
        closest = np.argsort(length)   #argsort the length             
                                                
        closey = [ train_np[a][3] for a in closest[:l]]
        vamsi = check(closey)    #checking closely and storing to vamsi
        if vamsi['W']<vamsi['M']:
            print('M')
        else:
            print('W')
```
